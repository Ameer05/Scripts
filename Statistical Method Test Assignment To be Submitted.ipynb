```python
#Statistical Test Assignment 
# Q4  An electronics firm is developing a new DVD player in response to customer requests. Using a prototype, the marketing team has collected focus data for different age groups viz. Under 25; 25-34; 35-44; 45-54; 55-64; 65 and above. Do you think that consumers of various ages rated the design differently?
# We will apply ANOVA Test method
dvd= pd.read_csv("dvdplayer.csv")
dvd
```

```python
# Actual Method using data frames where we have assigned numbers against the string values.
# H0:The consumers of various ages did not rated the design differently
# H1: The consumers of various ages rated the design differently
file_handler = open("dvdplayer.csv", "r") 
data = pd.read_csv(file_handler, sep = ",")
file_handler.close() 
```

```python
gegroup={'65 and over':1,'55-64':2,'Under 25':3,'25-34':4, '45-54':5, '35-44':6}

data.agegroup[data.agegroup == '65 and over'] = 1
data.agegroup[data.agegroup == '55-64'] = 2
data.agegroup[data.agegroup == 'Under 25'] = 3
data.agegroup[data.agegroup == '25-34'] = 4
data.agegroup[data.agegroup == '45-54'] = 5
data.agegroup[data.agegroup == '35-44'] = 6

print(data) 
```

```python
#Now using the above values (text and  numeric stored in the range 1 to 6)
s1= data[data['agegroup']== 1]
s2= data[data['agegroup']== 2]
s3= data[data['agegroup']== 3]
s4= data[data['agegroup']== 4]
s5= data[data['agegroup']== 5]
s6= data[data['agegroup']== 6]
 
print(s1,s2,s3,s4,s5,s6)
```

```python
# Now we will apply ANOVA 
stats.f_oneway(s1,s2,s3,s4,s5,s6) #pvalues is 0.0000308732491 <5%, hence, nullhypothesis is true i.e H0 is true.
```

```python
#Statistical test assignment Question no.5
# a. Is there any relationship in between labour force status with marital status? 
# H0: There is no reltionship, H1: There is a relationship

# Loading the data

Sample_Survey= pd.read_csv("sample_survey.csv")
Sample_Survey.head(5)
```

```python
# a. Is there any relationship in between labour force status with marital status?  Chisqaure method
t=pd.crosstab(Sample_Survey.wrkstat, Sample_Survey.marital , margins=True) #It will generate the observed output
t
```

```python
#Now we will genreate the expected values
stats.chi2_contingency(observed=t) #pvalue =-122.05180295
```
